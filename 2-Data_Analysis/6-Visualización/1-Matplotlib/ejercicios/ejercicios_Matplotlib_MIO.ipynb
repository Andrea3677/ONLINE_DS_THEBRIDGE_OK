2. Sabemos que podemos pintar gráficas de dos formas: la figura incluye los ejes o teniendo figura y ejes por separado.

Usando solo una figura, usa numpy para los valores del eje X entre 0 y 5. 

Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). 

Elige la precisión en el eje X que desees.  

```python
# Crear figura con 2 subplots (1 fila, 2 columnas)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Definir el eje X con precisión de 0.1
x = np.arange(0, 5.1, 0.1)

# Primera gráfica (izquierda): y = 3x (pendiente positiva por (0,0))
y1 = 3 * x
ax1.plot(x, y1, 'b-', linewidth=2, label='y = 3x')
ax1.set_title('Recta con Pendiente Positiva')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.grid(True, alpha=0.3)
ax1.legend()

# Segunda gráfica (derecha): y = -3x - 5 (pendiente negativa por (0,-5))
y2 = -3 * x - 5
ax2.plot(x, y2, 'r-', linewidth=2, label='y = -3x - 5')
ax2.set_title('Recta con Pendiente Negativa')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.grid(True, alpha=0.3)
ax2.legend()

# Ajustar el layout y mostrar
plt.tight_layout()
plt.show()
```

7. Crea un gráfico de barras. Pintamos la altura de un grupo de amigos, cada barra representa a una persona. Ana mide 160 cm, Luis mide 180 cm, Pedro mide 175 cm, Sofía mide 190 cm, Carmen mide 170 cm. Las barras serán verdes.

```python
# Datos de altura de los amigos
nombres = ['Ana', 'Luis', 'Pedro', 'Sofía', 'Carmen']
alturas = [160, 180, 175, 190, 170]
colores = ['green'] * len(nombres)  # Todas las barras verdes

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(nombres, alturas, color=colores, edgecolor='darkgreen', linewidth=1.2)

# Personalizar el gráfico
plt.title('Altura de los Amigos', fontsize=14, fontweight='bold', pad=20)
plt.xlabel('Amigos', fontsize=12)
plt.ylabel('Altura (cm)', fontsize=12)
plt.grid(axis='y', alpha=0.3, linestyle='--')

# Añadir los valores encima de las barras
for i, altura in enumerate(alturas):
    plt.text(i, altura + 2, f'{altura} cm', ha='center', va='bottom', fontsize=10)

# Ajustar los límites del eje Y para mejor visualización
plt.ylim(150, 200)

plt.tight_layout()
plt.show()
```

11. Prueba a invertir los ejes con ax.invert_axis()
Necesitarás tener los ejes disponibles fuera de la figura

```python
# Datos de altura de los amigos
nombres = ['Ana', 'Luis', 'Pedro', 'Sofía', 'Carmen']
alturas = [160, 180, 175, 190, 170]

# Crear figura y ejes por separado
fig, ax = plt.subplots(figsize=(10, 6))

# Crear el gráfico de barras horizontal
bars = ax.barh(nombres, alturas, color='green', edgecolor='darkgreen', linewidth=1.2)

# Personalizar el gráfico
ax.set_title('Altura de los Amigos (Eje Invertido)', fontsize=14, fontweight='bold', pad=20)
ax.set_xlabel('Altura (cm)', fontsize=12)
ax.set_ylabel('Amigos', fontsize=12)

# Eliminar la escala del eje X
ax.set_xticks([])

# INVERTIR EL EJE Y - Esto pondrá a Ana arriba y Carmen abajo
ax.invert_yaxis()

# Añadir los valores al final de las barras
for bar, altura in zip(bars, alturas):
    width = bar.get_width()
    ax.text(width + 2, bar.get_y() + bar.get_height()/2, 
            f'{altura} cm', ha='left', va='center', 
            fontsize=11, fontweight='bold')

# Ajustar límites para que los textos quepan bien
ax.set_xlim(150, 210)

# Cuadrícula tenue
ax.grid(axis='x', alpha=0.2, linestyle='-')

plt.tight_layout()
plt.show()
```

14. Ahora pintamos un gráfico de línea con las ventas respecto al año. Pon un título y etiquetas en los ejes.
Cambia el estilo a 'seaborn-white'. La línea debe ser con rayas y verde.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Aplicar el estilo 'seaborn-white'
plt.style.use('default')

# Crear el DataFrame
df = pd.DataFrame({
    'Year': [2015, 2016, 2017, 2018, 2019],
    'Sold_items_A': [1000, 3500, 4000, 5500, 7000]
})

# Crear el gráfico de línea
plt.figure(figsize=(10, 6))

# Gráfico de línea con rayas (linestyle='--') y color verde
plt.plot(df['Year'], df['Sold_items_A'], 
         color='green', 
         linestyle='--', 
         linewidth=2.5,
         marker='o',           # Añadir puntos en cada año
         markersize=8,
         markerfacecolor='darkgreen',
         markeredgecolor='black',
         markeredgewidth=1)

# Añadir título y etiquetas
plt.title('Evolución de Ventas - Producto A (2015-2019)', 
          fontsize=14, fontweight='bold', pad=20)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Items Vendidos', fontsize=12)

# Añadir cuadrícula para mejor lectura
plt.grid(True, alpha=0.3)

# Añadir los valores encima de los puntos
for year, sales in zip(df['Year'], df['Sold_items_A']):
    plt.text(year, sales + 150, f'{sales:,}', 
             ha='center', va='bottom', fontsize=10, fontweight='bold')

# Ajustar los límites del eje Y para que quepan los textos
plt.ylim(0, 8000)

# Formatear el eje Y para que muestre comas en miles
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x:,.0f}'))

plt.tight_layout()
plt.show()
```

15. Otro departamento B ha vendido en esos años 2000, 3100, 5000, 4000, 6000 unidades. Incluye esa columna en el DataFrame y pinta en la misma gráfica las dos líneas. B es una línea punteada y roja. Muestra una leyenda abajo a la derecha.

```python


# Usar un estilo disponible
plt.style.use('default')  # Puedes cambiarlo por cualquier estilo de plt.style.available

# Crear el DataFrame con ambas columnas de ventas
df = pd.DataFrame({
    'Year': [2015, 2016, 2017, 2018, 2019],
    'Sold_items_A': [1000, 3500, 4000, 5500, 7000],
    'Sold_items_B': [2000, 3100, 5000, 4000, 6000]
})

# Crear el gráfico de líneas
plt.figure(figsize=(10, 6))

# Línea para Departamento A (verde con rayas)
plt.plot(df['Year'], df['Sold_items_A'], 
         color='green', 
         linestyle='--', 
         linewidth=2.5,
         marker='o',
         markersize=6,
         markerfacecolor='darkgreen',
         markeredgecolor='black',
         markeredgewidth=1,
         label='Departamento A')

# Línea para Departamento B (roja punteada)
plt.plot(df['Year'], df['Sold_items_B'], 
         color='red', 
         linestyle=':',  # Línea punteada
         linewidth=2.5,
         marker='s',     # Marcador cuadrado
         markersize=6,
         markerfacecolor='red',
         markeredgecolor='black',
         markeredgewidth=1,
         label='Departamento B')

# Añadir título y etiquetas
plt.title('Evolución de Ventas - Departamento A vs B (2015-2019)', 
          fontsize=14, fontweight='bold', pad=20)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Items Vendidos', fontsize=12)

# Añadir cuadrícula
plt.grid(True, alpha=0.3)

# Añadir leyenda abajo a la derecha
plt.legend(loc='lower right', frameon=True, fancybox=True, shadow=True)

# Añadir los valores encima de los puntos
for year, sales_a, sales_b in zip(df['Year'], df['Sold_items_A'], df['Sold_items_B']):
    plt.text(year, sales_a + 200, f'{sales_a:,}', 
             ha='center', va='bottom', fontsize=9, fontweight='bold', color='green')
    plt.text(year, sales_b + 200, f'{sales_b:,}', 
             ha='center', va='bottom', fontsize=9, fontweight='bold', color='red')

# Ajustar los límites del eje Y para que quepan los textos
plt.ylim(0, 8000)

# Formatear el eje Y para que muestre comas en miles
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x:,.0f}'))

plt.tight_layout()
plt.show()
```

19. Vamos a pintar un histograma de unas alturas generadas aleatoriamente. Crea 300 alturas entre 150 y 190 cm.

Pista: https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.DataFrame.plot.html

```python

# Generar 300 alturas aleatorias entre 150 y 190 cm
alturas = np.random.uniform(150, 190, 300)

# Crear el histograma
plt.figure(figsize=(10, 6))
plt.hist(alturas, bins=15, color='skyblue', edgecolor='black', alpha=0.7)

# Personalizar el gráfico
plt.title('Distribución de 300 Alturas Aleatorias (150-190 cm)', fontsize=14, fontweight='bold')
plt.xlabel('Altura (cm)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.grid(axis='y', alpha=0.3)

# Añadir líneas de referencia para la media y desviación estándar
media = np.mean(alturas)
std = np.std(alturas)

plt.axvline(media, color='red', linestyle='--', linewidth=2, 
           label=f'Media: {media:.1f} cm')
plt.axvline(media + std, color='orange', linestyle=':', linewidth=1.5,
           label=f'+1σ: {media + std:.1f} cm')
plt.axvline(media - std, color='orange', linestyle=':', linewidth=1.5,
           label=f'-1σ: {media - std:.1f} cm')

plt.legend()
plt.tight_layout()
plt.show()

# Mostrar estadísticas descriptivas
print(f"Estadísticas de las alturas generadas:")
print(f"Media: {media:.2f} cm")
print(f"Desviación estándar: {std:.2f} cm")
print(f"Mínimo: {np.min(alturas):.2f} cm")
print(f"Máximo: {np.max(alturas):.2f} cm")
```
