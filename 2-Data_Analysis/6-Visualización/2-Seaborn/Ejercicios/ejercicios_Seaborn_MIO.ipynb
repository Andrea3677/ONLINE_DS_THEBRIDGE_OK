# 23. Mapas de calor
### Calcula las correlaciones del primer DataFrame original df
### Pista: las correlaciones se calculan con el método corr()
### Recuerda: las correlaciones miden entre dos variables, si varío una, cuánto me varía la otra (cómo están de relacionadas)

```python
# col_num = pokemon.select_dtypes(include='number').corr()

correlation_matrix = pokemon.select_dtypes(include='number').corr().corr()
sns.heatmap(correlation_matrix, 
            annot=True,        # Mostrar los valores numéricos
            cmap='coolwarm',   # Esquema de colores (azul-rojo)
            center=0,          # Centrar en 0
            square=True,       # Celdas cuadradas
            fmt='.2f',         # Formato de 2 decimales
            cbar_kws={'shrink': 0.8})  # Barra de color más pequeña
plt.tight_layout()
plt.show()
```

```python
# col_num = pokemon.select_dtypes(include='number').colums
# correlation_matrix = pokemon[col_num].corr()

correlation_matrix = pokemon[['Total', 'HP', 'Attack', 'Defense',
       'Sp. Atk', 'Sp. Def', 'Speed', 'Stage']].corr()
sns.heatmap(correlation_matrix, 
            annot=True,        # Mostrar los valores numéricos
            cmap='coolwarm',   # Esquema de colores (azul-rojo)
            center=0,          # Centrar en 0
            square=True,       # Celdas cuadradas
            fmt='.2f',         # Formato de 2 decimales
            cbar_kws={'shrink': 0.8})  # Barra de color más pequeña
plt.tight_layout()
plt.show()
```

# 26. Usa catplot con el DataFrame original.
### Muestra el valor de Defense según el 'Type 1', muestra un subplot para legendarios y otro para no legendarios y divide los colores según el Type 2

```python
grafico = sns.catplot(
    data=pokemon,
    x='Type 1', 
    y='Defense',
    hue='Type 2',           # Color por Type 2
    col='Legendary',        # Subplots separados por Legendario
    kind='box',             # Tipo de gráfico: box, violin, swarm, etc.
    palette='Set3',         # Paleta de colores
    height=6, 
    aspect=1.5,             # Relación ancho/alto
    sharex=True,            # Compartir eje X entre subplots
    legend_out=True         # Leyenda fuera del gráfico
)

# Personalizar el gráfico
grafico.set_axis_labels("Tipo Primario (Type 1)", "Puntos de Defensa")
grafico.set_titles("Pokémon {col_name}")  # Títulos automáticos para cada subplot
grafico.set_xticklabels(rotation=45)      # Rotar etiquetas del eje X

# Añadir título general
plt.subplots_adjust(top=0.85)
grafico.fig.suptitle('Distribución de Defensa por Tipo 1 - Separado por Legendarios y Type 2', 
               fontsize=16, fontweight='bold')

plt.tight_layout()
plt.show()
```
