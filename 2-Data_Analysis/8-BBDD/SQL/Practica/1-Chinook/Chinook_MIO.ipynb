```python
import pandas as pd
import sqlite3
```

```python
# Conectamos con la base de datos chinook.db
connection = sqlite3.connect("data/chinook.db")

# Obtenemos un cursor que utilizaremos para hacer las queries
crsr = connection.cursor()
```

```python
# Con esta función leemos los datos y lo pasamos a un DataFrame de Pandas
def sql_query(query):

    # Ejecuta la query
    crsr.execute(query)

    # Almacena los datos de la query 
    ans = crsr.fetchall()

    # Obtenemos los nombres de las columnas de la tabla
    names = [description[0] for description in crsr.description]

    return pd.DataFrame(ans,columns=names)
```

```python
# 1. Facturas de Clientes de Brasil, Nombre del cliente, Id de factura, fecha de la factura y el país de la factura

query = '''
SELECT  cus.FirstName, cus.LastName, i.invoiceid, i.invoicedate,i.billingcountry
FROM invoices as i
LEFT JOIN customers as cus
on i.customerid = cus.customerid
WHERE billingcountry = 'Brazil'

'''

sql_query(query)
```

```python
# 2. Obtén cada factura asociada a cada agente de ventas con su nombre completo

query = '''
SELECT  i.invoiceid, e.firstname, e.lastname
FROM invoices as i
LEFT JOIN customers as cus
on i.customerid = cus.customerid
LEFT JOIN employees as e
on cus.SupportRepId = e.EmployeeId

'''

sql_query(query)
```

```python
# 3. Obtén el nombre del cliente, el país, el nombre del agente y el total

query = '''
SELECT  cus.firstname, cus.lastname,cus.company,cus.country, e.firstname as rep_name, e.lastname as rep_lastname
from customers as cus
LEFT JOIN employees as e
on cus.SupportRepId = e.EmployeeId

'''

sql_query(query)
```

```python
# 4. Obtén cada artículo de la factura con el nombre de la canción
query = '''
SELECT  ii.trackid, ii.invoiceid, t.name
from invoice_items as ii
LEFT JOIN tracks as t
on ii.trackid = t.trackid

'''

sql_query(query)
```

```python
# 5. Muestra todas las canciones con su nombre, formato, álbum y género
query = '''
SELECT  t.name, mt.name, a.title, g.name
from tracks as t 
LEFT JOIN albums as a
on t.albumid = a.albumid
LEFT JOIN genres as g
on t.genreid = g.genreid
LEFT JOIN media_types as mt
on t.mediatypeid = mt.mediatypeid
'''

sql_query(query)
```

```python
# 6. Cuántas canciones hay en cada playlist
query = '''
SELECT p.name, p.playlistid, count(pt.trackid) as count_track
from playlists as p
LEFT JOIN playlist_track as pt
on p.playlistid = pt.playlistid
group by p.playlistid

'''

sql_query(query)
```

```python
# 7. Cuánto ha vendido cada empleado

query = '''
SELECT e.lastname, count(ii.trackid) as total_item
from customers as c
LEFT JOIN employees as e
on c.supportrepid = e.employeeid
LEFT JOIN invoices as i
on c.customerid = i.customerid
LEFT JOIN invoice_items as ii
on ii.invoiceid = i.invoiceid

'''

sql_query(query)
```

```python
# 7. Cuánto ha vendido cada empleado

query = '''
SELECT e.lastname, count(ii.trackid) as total_item, sum(i.total) as TOTAL
from customers as c
LEFT JOIN employees as e
on c.supportrepid = e.employeeid
LEFT JOIN invoices as i
on c.customerid = i.customerid
LEFT JOIN invoice_items as ii
on ii.invoiceid = i.invoiceid
GROUP BY e.firstname
'''

sql_query(query)
```

```python
# 8. ¿Quién ha sido el agente de ventas que más ha vendido en 2009?

query = '''
SELECT e.lastname, count(ii.trackid) as total_item, sum(ii.unitprice) as TOTAL
from customers as c
LEFT JOIN employees as e
on c.supportrepid = e.employeeid
LEFT JOIN invoices as i
on c.customerid = i.customerid
LEFT JOIN invoice_items as ii
on ii.invoiceid = i.invoiceid
WHERE strftime('%Y', invoicedate) = '2009'
GROUP BY e.firstname
order by total desc
limit 1
'''

sql_query(query)
```

```python
# 9. ¿Cuáles son los 3 grupos que más han vendido?

query = '''
SELECT a.name,sum(ii.unitprice*ii.quantity) as total
from artists as a
LEFT JOIN albums as al
on a.artistid = al.artist3id
LEFT JOIN tracks as t
on t.albumid = t.albumid
LEFT JOIN invoice_items as ii
on ii.trackid = t.trackid
LEFT JOIN invoices as i
on i.invoiceid = ii.invoiceid
GROUP BY a.name
ORDER BY total desc
limit 3
'''

sql_query(query)
```
