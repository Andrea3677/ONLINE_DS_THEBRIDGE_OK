### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv) and assign it to a variable called chipo.

### Step 3. Name of the max valued product

```python
chipo["item_price"] = chipo["item_price"].str.replace('$','')
chipo["item_price"] = pd.to_numeric(chipo["item_price"])
```

```python
#Agrego la columna unit price
chipo["unit_price"] = chipo["item_price"]/chipo["quantity"]
```

```python
chipo.describe(include= "all")
```

```python
chipo
```

```python
chipo.groupby("item_name")["unit_price"].min().max()
```

```python
chipo.groupby("item_name")["unit_price"].min().idxmax()
```

```python
tabla_agrupados = chipo.groupby("item_name")["unit_price"].min()
precio_max = tabla_agrupados.max()
productos_max = tabla_agrupados[tabla_agrupados == precio_max]
print(productos_max)
```
