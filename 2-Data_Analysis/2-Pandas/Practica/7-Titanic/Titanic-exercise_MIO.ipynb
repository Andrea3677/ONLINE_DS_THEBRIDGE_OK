24. Using [`.pivot_table()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html) to count how many women or men survived by class, call it `counted`.

Don't panic and read the documentation!




```python
counted = ps.pivot_table(titanic,values=['survived'],index=['pclass', 'gender'],aggfunc='sum')
counted
```

31. Using `matplotlib` find the appropriate visualization to show the distribution of the column `'age'`

```python
import matplotlib.pyplot as plt
```

```python

plt.figure(figsize=(10, 6)) # esto determina el tamaño del gráfico, está en pulgadas

# Crea un histograma
plt.hist(titanic['age'].dropna(), bins=30, edgecolor='black', alpha=0.4) # crea el histograma, se le aplica un el dropna para quitar los null 
#antes de graficar, edgecolor es el color de la barra y alpha es la transparencia de las barras (poner valores entre 0 y 1)
#los bins es la cantidad de barras que se vana a poner en el gráfico


#formato del gráfico
plt.title('Distribución de Edades en el Titanic', fontsize=14, fontweight='bold') # para establecer un titulo y su formato
plt.xlabel('Edad', fontsize=12) # Se le asigna el nombre al eje x 
plt.ylabel('Frecuencia', fontsize=12) # Se le asigna el nombre al eje y 
plt.grid(axis='y', alpha=0.3)

# para mostrar el gráfico
plt.tight_layout()#hace que los elementos no se solapen
plt.show() #renderiza la imagen del gráfico
```

32. Using `matplotlib` find the appropriate plot to visualize the column `'gender'`

```python
conteo_genero = titanic['gender'].value_counts()
conteo_genero
```

```python
conteo_genero.values
```

```python
conteo_genero.index
```

```python
# Contar los valores de la columna 'gender'
conteo_genero = titanic['gender'].value_counts()


plt.figure(figsize=(8, 6))
plt.bar(conteo_genero.index, conteo_genero.values, color=['lightblue', 'lightgreen'])#Genera un gráfico de barras verticales
#conteo_genero.index, establece las etiquetas del eje x
#Conteo_genero.values establece los valores en el eje y

plt.title('Distribución de Género en el Titanic', fontsize=14, fontweight='bold')
plt.xlabel('Género', fontsize=12)
plt.ylabel('Número de Pasajeros', fontsize=12)

# Añadir los valores encima de las barras
for i, valor in enumerate(conteo_genero.values):
    plt.text(i, valor + 10, str(valor), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Gráfico de pastel - para mostrar proporciones
plt.figure(figsize=(5, 5))
#Gener un grafico tipo pastel
plt.pie(conteo_genero.values, #Determina el tamaño de cada porción
        labels=conteo_genero.index, #determina las etiquetas de c/porción
        autopct='%1.1f%%', # para calcular los porcentajes 
        colors=['lightblue', 'lightgreen'],
        startangle=90)# marca el comienzo de la porción, 90 es desde arriba de todo.

plt.title('Distribución de Género en el Titanic', fontsize=14, fontweight='bold')
plt.axis('equal')  # Hace el círculo perfectamente redondo
plt.tight_layout()
plt.show()
```

34. Using [`kdeplot`]("https://seaborn.pydata.org/generated/seaborn.kdeplot.html") from `seaborn` represent those who not survived distributed by age.<>

Hint: First you should "filter" the `titanic` dataset where the column "survived" is 0, indexing the column `"age"` only.

Arguments you should pass to the function: 

    - color = "red"
    - label = "Not Survived"
    - shade = True
    
You should get something like this: 

![kdeplot](./img/kdeplotsur.png)

```python
dead_people_age = titanic['age'][titanic['survived'] == 0]

plt.figure(figsize=(5, 3))
# funcion para crear el gráfico de densidad (kdeplot)
sns.kdeplot(dead_people_age, 
            color='red', 
            label='Not Survived', 
            shade=True) # si se pone a False no se pinta el interior de la grafica

# Personalizar el gráfico
plt.title('Distribución de Edades de Pasajeros que No Sobrevivieron', fontsize=14)
plt.xlabel('Edad', fontsize=12)
plt.ylabel('Densidad', fontsize=12)
plt.legend() # muestra la leyenda
plt.grid(True, alpha=0.3) #activa la cuadrilla del fondo, si se pone a false no salen las lineas de atras, alpha es la transparencia de la lineas
plt.tight_layout()
plt.show()
```
