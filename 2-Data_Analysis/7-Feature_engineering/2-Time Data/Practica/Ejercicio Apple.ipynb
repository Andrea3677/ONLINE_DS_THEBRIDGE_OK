### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv)

```python
apple = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv')
```

### Step 3. Assign it to a variable apple and apply a head()

```python
apple.head()
```

### Step 4.  Check out the type of the columns

```python
apple.dtypes
```

### Step 5. Transform the Date column as a datetime type and check the info()

```python
apple['Date'] = pd.to_datetime(apple['Date'])
```

```python
apple.dtypes
```

```python
apple['Date']
```

### Step 6.  Set the date as the index

```python
apple.set_index('Date', inplace= True)
```

```python
apple
```

```python
apple.index
```

### Step 7.  Is there any duplicate dates?

```python
apple.index.has_duplicates
```

```python
apple.index.is_unique
```

### Step 8.  Ops...it seems the index is from the most recent date. Make the first entry the oldest date.

```python
apple.sort_index(inplace= True)
```

```python
apple
```

### Step 9. Get the last business day of each month

```python
pd.date_range(apple.index[0], apple.index[-1], freq= 'BM')
```

```python
apple.asfreq('BM')
```

### Step 10.  What is the difference in days between the first day and the oldest

```python
apple.index.max() - apple.index.min()
```

### Step 11.  How many months in the data we have?

```python
len(pd.date_range(apple.index[0], apple.index[-1], freq= 'M'))
```

```python
(apple.index[-1] - apple.index[0]).days / (365 / 12)
```

```python
apple.index.to_period('M')[0]-apple.index.to_period('M')[-1]
```
