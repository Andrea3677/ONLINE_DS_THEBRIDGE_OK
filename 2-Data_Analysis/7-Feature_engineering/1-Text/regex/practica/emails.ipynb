# Extrae los dominios de los correos con una expresi√≥n regular

```python
import re

emails = '''
miguellc@thebridgeschool.tech
miguel.cruz@uc3m.es
miguel-12-2000@mi-correo.net
miguellc@s@thebridgeschool.es
'''
```

Ejemplo:
* @thebridgeschool.tech
* @uc3m.es
* @mi-correo.net
* @thebridgeschool.es

```python
pattern = re.compile(r'@([a-zA-Z0-9.-]+([^@]\b))')

matches = pattern.finditer(emails)

for match in matches:
    print(match)
```

```python
pattern = re.compile(r'@([a-zA-Z0-9.-]+([^@]\b))')

matches = pattern.finditer(emails)

for match in matches:
    print(match)
```

```python
pattern = re.compile(r'@(\w+(?:-\w+)*+.\w+)')

matches = pattern.finditer(emails)

for match in matches:
    print(match)
```

```python
pattern = re.compile(r'@(\w+(?:-\w+)*+.\w+)')

matches = pattern.finditer(emails)

for match in matches:
    print(match)
```

```python
email = 'experto.en.regex@dominio.complicado.com'

pattern = re.compile(r'@([\w.]+)$')

matches = pattern.finditer(email)

for match in matches:
    print(match)
```
