## Juego del Ahorcado
El funcionamiento es que vaya mostrando por pantalla ```_ _ _ _ _``` las rayas bajas correspondientes al número de letras que tiene la palabra a adivinar. El usuario deberá insertar una letra y, si esa letra existe, se mostrará o mostrarán las letras acertadas ```S _ S _ _```. 

Si falla, se mostrará el número de errores que lleva y, si lleva 6 errores, entonces se mostrará "Fin del juego".
Encapsúlalo en una función.

````python
#FUNCIONES:

#Funcion para generar la base de los guiones de la palabra adivinar
def generar_base(palabra_adivinar):
    base = []
    for letra in palabra_adivinar:
        base.append(" __ ")
    return base

#Función para saber si una letra esta dentro de la palabra
def comprobar_intento(palabra_adivinar, base, intento_letra):
    for indice, letra in enumerate(palabra_adivinar):
        if letra == intento_letra:
            base[indice] = intento_letra.upper()
        else:
            continue
    return base

#Función para imprimir la base:
def imprimir_base(base):
    return " ".join(base)

#Función comprobar si la partida continua
def comprobar_continua(vidas, base):
    if vidas == 0:
        partida = False
        print(f"Perdiste, te quedaste sin vidas. La palabra era {palabra_adivinar}")
        return partida
    elif " __ " not in base:
        partida = False
        print(f"Ganaste. Palabra completa : {" ".join(base)}")
        return partida
    else:
        partida = True 
        print(" ".join(base))
        return partida

    
````

````python
#Codigo sin funciones aplicadas o con muy poquitas
vidas = 6
 #Se pregunta la palabra a adivinar
palabra_adivinar = input("Indique la palabra a adivinar").lower()
#Se genera las lineas de los lugares de las letras
base = generar_base(palabra_adivinar)
print(base)
partida = True

while partida == True:
    intento_letra = input("Indica una letra").lower()
    #Se comprueba si la letra esta o no en la palabra a adivinar y si esta se reemplaza en el tablero de las las lineas, si no esta se reta una vida
    if intento_letra in palabra_adivinar:
        for indice, letra in enumerate(palabra_adivinar):
            if letra == intento_letra:
                print("Acertaste")
                base[indice] = intento_letra.upper()
            else:
                continue
    else: 
        vidas -= 1
        print(f"No acertaste. Te quedan {vidas} vidas")
        

#Se comprueba si la partida debe continuar o no. 
    if vidas == 0:
        partida = False
        print(f"Perdiste, te quedaste sin vidas. La palabra era {palabra_adivinar}")
    elif " __ " not in base:
        partida = False
        print(f"Ganaste. Palabra completa : {" ".join(base)}")
    else:
        partida = True 
        print(" ".join(base))
````

````python
#Código con las funciones aplicadas
vidas = 6
palabra_adivinar = input("Indique la palabra a adivinar").lower()
base = generar_base(palabra_adivinar)
print(base)
partida = True

while partida == True:
    intento_letra = input("Indica una letra").lower()
    if intento_letra in palabra_adivinar:
        comprobar_intento(palabra_adivinar, base, intento_letra)
    else: 
        vidas -= 1
        print(f"No acertaste. Te quedan {vidas} vidas")
    partida = comprobar_continua(vidas, base)
````
