    * Eliminar libro: se solicita al usuario el título del libro y éste debe ser eliminado de la lista de libros
    * Alquilar libro: se solicita al usuario el título del libro y se debe pasar el campo "Alquilado" a True

```python
#PREGUNTAR AL USUARIO QUE QUIERE HACER
funcion = input("Elige entre las siguientes opciones:" \
"1 - Visualizar todos los libros de la biblioteca"
"2 - Buscar libro" \
"3 - Añadir libro"
"4 - Eliminar libro por título" \
"5 - Alquiler libro" \
"6 - Devolver libro ")

#VARIABLES BUSCAR LIBRO
libro_encontrado = None


if funcion == "2":
    libro_buscar = input("Escribe el titulo del libro")
elif funcion == "3":
    titulo_agregar = input ("Escribe el titulo a añadir")
    autor_agregar = input(f"Indica el autor de {titulo_agregar}")
elif funcion == "4":
    titulo_eliminar = input ("Escribe el titulo a añadir")
elif funcion == "5":
    titulo_alquilar = input("Escribe el titulo del libro que quieres alquilar")
elif funcion == "6":
    titulo_devolver = input("Escribe el titulo del libro que quieres devolver")
else:
    print("Entrada inválida. Por favor, vuelva a intentarlo.")


match funcion:
    case "1":
        for elemento in libros:
            print(elemento)
    case "2":
        for elemento in libros:
            if elemento["Titulo"] == libro_buscar:
                print(elemento)
                libro_encontrado = True
                break
            else:
                libro_encontrado = False
                continue 
        if libro_encontrado == False:
            print("No tenemos ese libro")
    case "3":
        libros.append({"Titulo": titulo_agregar, "Autor": autor_agregar, "Alquilado": False})
        print(f"Libro añadido. {libros[-1]}")
    case "4": 
        for indice, elemento in enumerate(libros):
            if elemento["Titulo"] == titulo_eliminar:
                libros.pop(indice)
                print(f"{titulo_eliminar} ha sido eliminado")
                libro_encontrado = True
                break
            else:
                libro_encontrado = False
                continue
        if libro_encontrado == False :
            print(f"No tenemos ningun libro llamado {titulo_eliminar}. No se ha eliminado ningun libro")
    case "5":
        for elemento in libros:
            if elemento["Titulo"] == titulo_alquilar:
                if elemento["Alquilado"] == True:
                    print(f"{titulo_alquilar} no está disponible en estos momentos, intentelo en los próximos días.")
                    libro_encontrado = True
                else:
                    elemento["Alquilado"] = True
                    print(f"Puede llevarse {titulo_alquilar}. Que disfrute de la lectura.")
                    libro_encontrado = True
                break
            else:
                libro_encontrado = False
                continue
        if libro_encontrado == False :
            print(f"No tenemos ningun libro llamado {titulo_eliminar}.")
    case "6":
        for elemento in libros:
            if elemento["Titulo"] == titulo_alquilar:
                elemento["Alquilado"] = False
                print(f"Devolución registrada. Muchas gracias.")
                libro_encontrado = True
                break
            else:
                libro_encontrado = False
                continue
        if libro_encontrado == False :
            print(f"No encontramos ese libro en nuestros registro, corrobore que haya escrito bien el título.")

    
```
