```python
#Funci√≥n para crear tablero en blanco
def crear_tablero(lados):
    tablero=[]
    for elemento in range(0,lados,1):
        fila = []
        for elemento in range (0,lados,1):
            fila.append(" ")
            tablero.append(fila)
    return tablero

```
