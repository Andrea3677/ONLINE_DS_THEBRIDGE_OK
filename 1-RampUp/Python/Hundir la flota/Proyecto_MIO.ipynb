```python
import random

#Función para crear tablero en blanco
def crear_tablero(lados):
    tablero=[]
    for elemento in range(0,lados,1):
        fila = []
        for elemento in range (0,lados,1):
            fila.append(" ")
        tablero.append(fila)
    return tablero

#Funcion para imprimir tablero
def imprimir_tablero(tablero):
    tamaño = len(tablero)+1
    # Encabezado con números
    print("    " + "   ".join(str(i) for i in range(1,tamaño)))
    # Filas con números
    for i, fila in enumerate(tablero):
        print(f"{i} | " + " | ".join(fila))

    print()

#Función para pintar los barcos en el tablero
def pintar_barcos(barcos,tablero):
    for valor in barcos.values():
        for elemento in valor:
            i = elemento[0] - 1
            j = elemento[1] - 1
            tablero[i][j] = "B"
    return tablero


#Generar las posiciones de los barcos

# #Variables barco para el jugador humano
# barcos_jh = {
# "barco4_1" : [[9,2],[9,3],[9,4],[9,5]],
# "barco3_1" : [[1,1],[2,1],[3,1]],
# "barco3_2" : [[5,3],[5,4],[5,5]],
# "barco2_1" : [[2,4],[2,5]],
# "barco2_2" : [[5,7],[6,7]],
# "barco2_3" : [[6,9],[7,9]],
# "barco1_1" : [[7,4]],
# "barco1_2" : [[2,8]],
# "barco1_3" : [[10,8]],
# "barco1_4" : [[4,10]],
# }

largo_barcos = {
        "barco4_1": 4,
        "barco3_1": 3,
        "barco3_2": 3,
        "barco2_1": 2,
        "barco2_2": 2,
        "barco2_3": 2,
        "barco1_1": 1,
        "barco1_2": 1,
        "barco1_3": 1,
        "barco1_4": 1
    }


#Función para pintar los barcos en el tablero
def pintar_barcos(barcos,tablero):
    for valor in barcos.values():
        for elemento in valor:
            i = elemento[0] - 1
            j = elemento[1] - 1
            tablero[i][j] = "B"
    return tablero

#Se crea el tablero de disparos a fin de registrar los disparos del jugador humano
tablero_disparos = crear_tablero(10)
tablero_jugador = crear_tablero(10)
tablero_maquina = crear_tablero(10)
```
