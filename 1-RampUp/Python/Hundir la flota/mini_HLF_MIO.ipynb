## Mini Hundir la Flota
#### Primeros pasos
Vamos a empezar a programar trozos de código que nos servirán para el HLF:

* Crear un Tablero como una lista de listas de 10x10 que contenga caracteres espacio: " "
* Insertar dos barcos sustituyendo en el Tablero los espacios en blanco por el caracter "B":
    * Uno de 4 casillas en horizontal: casillas [1][0], [1][1], [1][2], [1][3]
    * Uno de 3 casillas en vertical:   casillas [3][3], [4][3], [4][4], [4][5]
* Escribir una celda de código separada de las anteriores para simular que el usuario está efectuando un disparo: 
    *  Solicite al usuario dos coordenadas `i,j` mediante dos inputs
    *  Compruebe si esas coordenadas están ocupada por una casilla barco (que haya una en "B")
        * Si hay una "B":
            * Cambiar el contenido de esa posición por una "x"
            * Imprimir por pantalla el texto "Tocado en posición `i,j`"
        * Si hay un " ":
            * Cambiar el contenido de esa posición por una "o"
            * Imprimir por pantalla el texto "Agua"
    * Imprimir por pantalla el Tablero
* Para comprobar que funciona bien la celda de disparar ejecutarla varias veces metiendo distintas coordenadas

### Bonus
* En la celda de disparar, si el usuario acierta a tocar un barco, vuelve a disparar
* Resuélvelo con funciones:
    * Una para solicitar una coordenada válida a la que llamarás dos veces, una par i y otra par j
    * Una `disparar` que devuelva `True` si el usuario ha acertado o `False` si no.
    


```python
#Código que genera un tablero en blanco
tablero =[]

for elemento in range(0,10,1):
    fila = []
    for elemento in range (0,10,1):
       fila.append(" ")
    tablero.append(fila)
print(tablero)
```

```python
# Función para colocar los barcos en el tablero
def ubicar_barco(*args):
    for elemento in args:
        x = elemento[0]
        y = elemento[1]
        tablero[x][y] = "B"
    return tablero


#Función para los inputs
def input_coordenada():
    coordenada_i = int(input("Indica la coordenada i"))-1
    coordenada_j = int(input("Indica la coordenada j"))-1
    return coordenada_i, coordenada_j

#Funcion para segundo disparo 
def disparar(coordenada_i, coordenada_j):
    if tablero[coordenada_i][coordenada_j] == "B":
        return True
    elif tablero[coordenada_i][coordenada_j] == " ":
        return False
    else:
        return False
```

```python
#ubicamos los barcos
ubicar_barco([1,0], [1,1], [1,2], [1,3])
ubicar_barco([3,3], [4,3], [5,3])
```

```python
#Se pide al usurio las coordenadas de donde quiere atacar y se comprueba en el tablero dicha posición 

segundo_disparo = True
try:

    while segundo_disparo == True:
        coordenada_i, coordenada_j = input_coordenada()
        if coordenada_i >= 10 or coordenada_j >= 10:
             print(f"Coordenada {coordenada_i + 1},{coordenada_j + 1}, inválida. Error: Coordenada fuera del límite del tablero")
             break
        else:
            segundo_disparo = disparar(coordenada_i,coordenada_j)
            if tablero[coordenada_i][coordenada_j] == "B":
                tablero[coordenada_i][coordenada_j] = "X"
                print(f"Tocado en posición {coordenada_i + 1},{coordenada_j + 1}")
            elif tablero[coordenada_i][coordenada_j] == " ":
                tablero[coordenada_i][coordenada_j] = "O"
                print(f"Agua en posición {coordenada_i + 1},{coordenada_j + 1}")
            elif tablero[coordenada_i][coordenada_j] == "X" or tablero[coordenada_i][coordenada_j] == "O":
                print(f"Posición {coordenada_i + 1},{coordenada_j + 1} ya había sido disparada")
            else:
                print(f"Coordenada {coordenada_i + 1},{coordenada_j + 1}, inválida.")

    for elemento in tablero:
            print(elemento)

except Exception as e: print(e)

```
